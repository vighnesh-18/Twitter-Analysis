Convert the target "4" to "1"

```python
twitter_data.replace({'target':{4:1}},inplace=True)
twitter_data['target'].value_counts()
```

0--> -ve tweet
1--> +ve tweet

**Stemming**

Stemming is a process of reducing a word to it's root word

ex: actor,actress,acting = act

```python
port_stem= PorterStemmer()
```

```python
def stemming(content) :
    stemmed_content=re.sub('[^a-zA-Z]',' ',content)
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)

    return stemmed_content
```

```python
twitter_data['stemmed_content']=twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
#separating data and label
X=twitter_data['stemmed_content'].values
Y=twitter_data['target'].values
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data and test data

Test size=0.2 -> 80% training and 20% testing

Stratify --> in 20% of testing data it will make sure that it contains equal 0'1 ans 1's
same for training data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

converting the textual data to numerical data

```python
vectorizer=TfidfVectorizer()

X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

Training the Machine Learning Model

Logistic Regression

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

model evaluation &
Accuracy score

```python
#accuracy score on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on the training data:',training_data_accuracy)
```

```python
#accuracy score on testing data
X_train_prediction=model.predict(X_test)
training_data_accuracy=accuracy_score(X_train_prediction,Y_test)
```

```python
print('Accuracy score on test data:',training_data_accuracy)
```

Model accuracy=77.8%

saving the trained model

```python
import pickle
filename='trained_model.sav'
pickle.dump(model,open(filename,'wb'))
```
